```python
import requests
from dotenv import load_dotenv
import os

load_dotenv()
# --- CONFIGURATION ---
URL = "https://ai-foundry-demo79-resource.services.ai.azure.com/api/projects/ai-foundry-demo79/applications/SupportBot-AK/protocols/openai/responses"
API_KEY = os.getenv('FOUNDARY_API')
API_VERSION = "2025-11-15-preview"

system_instruction = """
You are an Interview Simulator. 
STRICT CYCLE:
1. Ask ONE question.
2. Wait for user input.
3. Provide an 'Analysis Report' (Feedback, Score 1-10, Tip).
4. Ask the NEXT question.
Only output the Interview content. Do not output 'Interviewer is analyzing' or other status text.
"""

history = [{"role": "system", "content": system_instruction}]

def get_agent_response(user_input):
    global history
    if user_input:
        history.append({"role": "user", "content": user_input})
    
    headers = {"api-key": API_KEY, "Content-Type": "application/json"}
    params = {"api-version": API_VERSION}
    payload = {"input": history}

    try:
        response = requests.post(URL, headers=headers, json=payload, params=params)
        data = response.json()
        
        for item in data.get('output', []):
            if item.get('type') == 'message':
                text = item['content'][0]['text']
                history.append({"role": "assistant", "content": text})
                return text
    except Exception as e:
        return f"Error: {e}"

# --- CLEAN FLOW ---
print("--- Interview Simulator Ready ---")
# Start with an empty user input to trigger the first question
print(f"\nInterviewer: {get_agent_response(None)}")

while True:
    user_answer = input("\nCandidate: ")
    if user_answer.lower() == 'exit':
        break
    
    # This will print the Report AND the Next Question
    print(f"\n{get_agent_response(user_answer)}")
```
