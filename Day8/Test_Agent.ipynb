```python
import requests
import os
from dotenv import load_dotenv

load_dotenv()
# --- CONFIGURATION ---
# Use your stable PUBLISHED application endpoint
URL = "https://ai-foundry-demo79-resource.services.ai.azure.com/api/projects/ai-foundry-demo79/applications/email-assistant/protocols/openai/responses"
API_KEY = os.getenv('FOUNDARY_API')
API_VERSION = "2025-11-15-preview"

def ask_agent(user_query):
    headers = {
        "api-key": API_KEY,
        "Content-Type": "application/json"
    }
    params = {"api-version": API_VERSION}
    
    # Payload structure for the Responses API
    payload = {
        "input": [{"role": "user", "content": user_query}]
    }

    try:
        response = requests.post(URL, headers=headers, json=payload, params=params)
        
        if response.status_code == 200:
            # CORRECTED: Parse the response into a dictionary
            data = response.json()
            
            # Drill down to the agent's text response
            # Path: response -> output (list) -> content (list) -> text
            return data['output'][0]['content'][0]['text']
        else:
            return f"Error {response.status_code}: {response.text}"
            
    except Exception as e:
        return f"An error occurred: {str(e)}"

# --- INTERACTIVE LOOP ---
print("--- Email Assistant Initialized (Type 'exit' to quit) ---")

while True:
    user_input = input("\nYou: ")
    
    if user_input.lower() == 'exit':
        print("Goodbye!")
        break
    
    
    print("Agent is thinking...")
    result = ask_agent(user_input)
    print(f"Agent: {result}")
```
