```python
import requests
import os
from dotenv import load_dotenv


load_dotenv()
URL = "https://ai-foundry-demo79-resource.services.ai.azure.com/api/projects/ai-foundry-demo79/applications/SupportBot-AK/protocols/openai/responses"
API_KEY = os.getenv('FOUNDARY_API')
API_VERSION = "2025-11-15-preview"

def ask_agent(user_query):
    headers = {
        "api-key": API_KEY,
        "Content-Type": "application/json"
    }
    params = {"api-version": API_VERSION}
    payload = {"input": [{"role": "user", "content": user_query}]}

    try:
        response = requests.post(URL, headers=headers, json=payload, params=params)
        if response.status_code != 200:
            return f"Error {response.status_code}: {response.text}"
            
        data = response.json()
        
        for item in data.get('output', []):
            if item.get('type') == 'message':
                return item['content'][0]['text']
        
        return "Agent responded, but no message text was found."
            
    except Exception as e:
        return f"Parsing Error: {str(e)}"

# --- INTERACTIVE LOOP ---
print("--- SupportBot-AK Initialized (Type 'exit' to quit) ---")

while True:
    user_input = input("\nCustomer: ")
    
    if user_input.lower() == 'exit':
        print("Goodbye!")
        break
    
    print("Bot is thinking...")
    result = ask_agent(user_input)
    print(f"SupportBot: {result}")
```
